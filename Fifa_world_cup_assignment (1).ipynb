```python
#Number of goal scored country from 1930 to 2014
Home=df_matches[['Home Team Name', 'Home Team Goals']].dropna()
Away=df_matches[['Away Team Name', 'Away Team Goals']].dropna()
```

```python
#taking only coulmns countries and goals
Home.columns = ['Countries', 'Goals']
Away.columns = Home.columns
```

```python
#appending both Home and away
goals = Home.append(Away, ignore_index=True)
```

```python
#now we finding the number goals score by each country from 1930 to 2014
goals = goals.groupby('Countries').sum()
Goals=goals.sort_values(by='Goals',ascending=False)
Goals
```

```python
# we are taking only top 20 by goals
plt.figure(figsize=(10,8))
plt.title('Countries that hit most no of goals',size=15,weight='bold')
ax=sns.barplot(data=Goals[:20],x=Goals[:20].index,y='Goals')
plt.xticks(rotation=45)
ax.bar_label(ax.containers[0])
plt.show()
```

We find that if the goal is more your chance of winning world cup is more.

```python
# We need to find the goal score by teams at each year
home = df_matches[['Year','Home Team Name', 'Home Team Goals']].dropna()
away = df_matches[['Year','Away Team Name', 'Away Team Goals']].dropna()
```

```python
# we include only three columns
home.columns = ['Year', 'Countries', 'Goals']
away.columns = home.columns
```

```python
#appending both dataframe home and away
year_goals = home.append(away, ignore_index=True)
```

```python
#creating new dataframe
year_goals = year_goals.groupby(['Year','Countries']).sum().reset_index()
year_goals
```

```python
#Now we know the goal score by each participated country in each year 
#Now we merged with df_worldcup data frame so that we know that how much goal scored by winner country
year_winner_goal=df_worldcup.merge(year_goals,left_on=["Year", "Winner"],right_on=["Year", "Countries"],how="left")
```

```python
#now we get dataframe which give th
year_winner_goal
```

```python
year_winner_goal['Runners-Up']=year_winner_goal['Runners-Up'].str.replace('Germany FR','Germany')
```

```python
#Now we rename the goal column to winner_goals
year_winner_goal.rename(columns={'Goals':'Winner_goals'},inplace=True)
```

```python
#Now we want to know to winner scored by runner_up country
year_winner_runner_up_goals = year_winner_goal.merge(year_goals, left_on=["Year", "Runners-Up"], right_on=["Year", "Countries"], how="left")
```

```python
year_winner_runner_up_goals
```

```python
year_winner_runner_up_goals.rename(columns={'Goals':'runner_up_goals'},inplace=True)
```

```python
year_winner_runner_up_goals
```

```python
year_winner_runner_up_goals['Third']=year_winner_runner_up_goals['Third'].str.replace('Germany FR','Germany')
```

```python
#Now we also want to third team also who came third in tournament
year_top_3_goals = year_winner_runner_up_goals.merge(year_goals, left_on=["Year", "Third"], right_on=["Year", "Countries"], how="left")
```

```python
year_top_3_goals
```

```python
#Now rename goals to third goals
year_top_3_goals.rename(columns={'Goals': 'Third_Goals'},inplace = True)
```

```python
#Now we ploting winner,runner_up and third team goal for each world cup
fig, ax = plt.subplots(figsize=(20,8))

plt.title('Number of Goals between top 3 each year',size=20,weight='bold')

columns = ['Winner_goals','runner_up_goals','Third_Goals']

plot_data = pd.melt(year_top_3_goals, id_vars='Year', value_vars=columns)

sns.barplot(x="Year", y="value", hue='variable', data=plot_data, linewidth=2.5, edgecolor=".2", ax=ax)

plt.tick_params(labelleft=False, left=False,labelsize=14)
ax.set_ylabel("Number of Goals")
ax.set_xlabel("Year")
plt.xticks(rotation=45)
for container in ax.containers:
    ax.bar_label(container)
```

```python
# add a new column that shows the column having the greater value
df_matches["Winner_match"] = df_matches[["Home Team Goals","Away Team Goals"]].idxmax(axis=1)
df_matches["Winner_match"] = df_matches["Winner_match"].str.replace("Home Team Goals", 'First').str.replace("Away Team Goals", 'Second')
```

```python
#now we checking the dataframe head
df_matches.head()
```

```python
# Now we adding one more column to see who is the winner of half_time
df_matches["Winner_half_time"] = df_matches[["Half-time Home Goals","Half-time Away Goals"]].idxmax(axis=1)
df_matches["Winner_half_time"] = df_matches["Winner_half_time"].str.replace("Half-time Home Goals", 'First').str.replace("Half-time Away Goals", 'Second')
```

```python
#checking the head
df_matches.head()
```

```python
# add a new column comparing 2 columns
df_matches["Win_half_time_win_all"] = df_matches["Winner_match"]==df_matches["Winner_half_time"]
```

```python
#now we checking the head once again
df_matches.head()
```

```python
#as our dataframe is very big we cannot see by only see for that we use value count funcation to see exact result
df_matches["Win_half_time_win_all"].value_counts()
```

There is high chance to win the match if the team win the first half.

```python
# group the data by year and calculate the percentage of true values
grouped = df_matches.groupby('Year')['Win_half_time_win_all'].mean() * 100
grouped = grouped.round(2)

# plot the percentage of true values for each year as a bar chart
plt.plot(grouped.index, grouped.values)
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Percentage of Win_half_time_win_all by Year')
plt.ylim([0, 100])
plt.show()
```

from the graph we have seen that winning the first half can increase your 80 percent to win the game
